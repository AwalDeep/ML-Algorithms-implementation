```python
features=2
sample=1000
x_data=np.random.rand(sample,features)
ones_array=np.ones(sample).reshape(sample,1)
x_final=np.append(ones_array,x_data,axis=1)
```

#random Y
y_true=np.random.choice([0,1],size=sample).reshape(sample,1)

```python
#Sigmoid Y
noise=np.random.randn(sample,1)
y_true=np.ones((sample,1))
betas=np.random.rand(features+1,1)
z0=np.dot(x_final,betas)+noise
e_z0=np.exp(-z0)
for i in range(len(z0)):
    y_true[i]=1/(1 +e_z0[i])
y_true=np.round(y_true)
y_true
```

```python
betas=np.random.rand(features+1,1)
one_minus_y_true=ones_array-y_true
#betas.shape
#x_final.shape
z=np.dot(x_final,betas)
```

```python
for i in range(1000):
    y_hat=np.ones((sample,1))

    e_z=np.exp(-z)
    #print(e_z)
    
    for i in range(len(e_z)):
        y_hat[i]= 1/(1 +e_z[i])
        
    
    one_minus_y_hat=ones_array-y_hat
    derivative=(np.dot(np.transpose(x_final),y_true-y_hat))/sample
    betas=betas+0.1*derivative
    max_log_likelihood=-(y_true*np.log(y_hat) + one_minus_y_true*np.log(one_minus_y_hat))
    z=np.dot(x_final,betas)
print(np.mean(max_log_likelihood))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
classifier = LogisticRegression(random_state = 0)
y_true=y_true.ravel().astype(int)
classifier.fit(x_data,y_true)
y_pred=classifier.predict_proba(x_data)
print(f"Intercept is {classifier.intercept_}")
print(f"Other betas are {classifier.coef_}")
print(f"Max.log likelihood is {log_loss(y_true, y_pred)}")
```
